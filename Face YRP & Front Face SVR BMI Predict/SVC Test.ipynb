```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## 讀取資料

```python
raw_df = pd.read_csv('svc50.csv')
raw_df.head(5)
```

```python
raw_data = list(raw_df.values)
```

## 抽取 test-data，並製作 test-df

```python
test_size = 0.2
test_count = int(len(raw_df) * test_size)
pick_index = list(np.random.choice(len(raw_df), test_count, replace=False))
```

```python
test_data = list()
for i in pick_index:
    test_data.append(raw_data.pop(i))
```

```python
test_df = pd.DataFrame(columns=raw_df.columns)
for d in test_data:
    test_df.loc[len(test_df)] = d
    
test_df
```

## 製作 train-df

```python
train_df = pd.DataFrame(columns=raw_df.columns)
for d in raw_data:
    train_df.loc[len(train_df)] = d

train_df
```

```python
x_train = train_df.iloc[:,5:len(train_df.columns)].values
x_train
```

```python
y_train = train_df.iloc[:,4:5].values
y_train = np.hstack(y_train)
y_train
```

```python
x_test = test_df.iloc[:,5:len(train_df.columns)].values
x_test
```

```python
y_test = test_df.iloc[:,4:5].values
y_test = np.hstack(y_test)
y_test
```

## SVC

```python
clf = SVC()
```
